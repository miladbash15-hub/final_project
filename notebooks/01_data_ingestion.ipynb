```
import pandas as pd
aisles=pd.read_csv('/content/aisles.csv')
departments= pd.read_csv('/content/departments.csv')
order_products__prior=pd.read_csv('/content/order_products__prior.csv')
order_products__train=pd.read_csv('/content/order_products__train.csv')
orders=pd.read_csv('/content/orders.csv')
products=pd.read_csv('/content/products.csv')
print(departments.head())
```

```
aisles = aisles.astype({
    "aisle_id": "int16",
    "aisle":"category"
})
departments = departments.astype({
    "department_id": "int16",
    "department": "category"
})
order_products__prior = order_products__prior.astype({
    "order_id":"int32",
    "product_id":"int32",
    "add_to_cart_order":"int16",
    "reordered":"int8"
})
order_products__train=order_products__train.astype({
    "order_id":"int32",
    "product_id":"int32",
    "add_to_cart_order":"int16",
    "reordered":"int8"

})
orders=orders.astype({
    "order_id":"int32",
    "user_id":"int32",
    "eval_set":"category",
    "order_number":"int16",
    "order_dow":"int8",
    "order_hour_of_day":"int8"
})
products=products.astype({
    "product_id":"int32",
    "aisle_id":"int16",
    "department_id":"int16",
    "product_name":"category"

})
```

```
products_full=(
    products
    .merge(aisles,on="aisle_id", how="left")
    .merge(departments,on="department_id", how="left")
)
```

```
prior_full= (
    order_products_prior
    .merge(orders, on="order_id", how="left")
    .merge(products_full, on="product_id", how="left")
)
```

```
train_full=(
    order_products__train
    .merge(orders, on="order_id", how="left")
    .merge(products_full,on="product_id",how="left")
)
train_full.info()
```

```
import pandas as pd

datasets= {
    "Prior Orders": prior_full,
    "Train Orders": train_full,
    "Products": products_full,
}
for name,  df in datasets.items():
  print(f"\n--- Missing Values in {name}---")
  missing = df.isna().mean().sort_values(ascending=False)
  print(missing)
```

```
order_products_full = pd.concat(
    [train_full,prior_full],
    axis=0,
    ignore_index=True
)
```

```
order_products_full = order_products_full.sample(face=0.04,random_state=42)
```
